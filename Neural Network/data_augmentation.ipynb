## resize Picture

```python
for name in os.listdir(path):
    img = cv2.imread(path+name)
    cropped = img[750:4350,1400:5000] #Y,X
    img = cv2.resize(cropped, (2400,2400))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imwrite('./SKR-HEX-20221116/cropped/'+name,img)
```

# flip pic

```python
path='./SKR-HEX-20221116/cropped/'
for name in os.listdir(path):
    img = cv2.imread(path+name)
    img_flip_along_x = cv2.flip(img, 0)
    img_flip_along_y = cv2.flip(img, 1)
    img_flip_along_xy = cv2.flip(img, -1)
    cv2.imwrite('./SKR-HEX-20221116/cropped/flip/x'+name,img_flip_along_x)
    cv2.imwrite('./SKR-HEX-20221116/cropped/flip/y'+name,img_flip_along_y)
    cv2.imwrite('./SKR-HEX-20221116/cropped/flip/xy'+name,img_flip_along_xy)
    
```
