```python
import numpy as np  
import pandas as pd  
import statsmodels.api as sm 
import os
import re
import openpyxl
```

```python
file = 'TC-1.csv'
filename=os.listdir()
output=pd.DataFrame(columns=['name','constant','coef1','coef2','Adj.R-2'])
for fn in filename:
    if re.search(r'^TC-', fn) and  re.search(r'\.csv$', fn):
        data = pd.read_csv(fn,header=1) 
        data.columns=['step','time','load','strainmean','strain1','strain2']
        df20=data[data['load']>20] #DATAover20kN
        x = sm.add_constant(df20.iloc[:,4:]) #Generate independent variables 
        y = df20['load'] #Generate dependent variable   因变量
        model = sm.OLS(y, x) #generative model
        result = model.fit() #model fit  
        #result.summary() #Model description 
        output.loc[len(output)]=(fn,result.params[0],result.params[1],result.params[2],result.rsquared_adj)
```

```python
output.to_csv('TC_output.csv', sep='\t', encoding='utf-8')
output.to_excel("TC_output.xlsx")  
output
```
