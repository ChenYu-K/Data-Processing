```python
file = 'no-1.csv'
filename=os.listdir()
output=pd.DataFrame(columns=['name','constant','coef','Adj.R-2'])
for fn in filename:
    if re.search(r'^no-', fn) and  re.search(r'\.csv$', fn):
        data = pd.read_csv(fn,header=1, encoding='shift-jis') 
        data.columns=['step','time','load','disp','strain1','strain2','na','strainmean']
        df20=data[data['load']>20] #DATAover20kN
        x = sm.add_constant(df20.iloc[:,7]) #Generate independent variables 
        y = df20['load'] #Generate dependent variable   因变量
        model = sm.OLS(y, x) #generative model
        result = model.fit() #model fit  
        #result.summary() #Model description 
        output.loc[len(output)]=(fn,result.params[0],result.params[1],result.rsquared_adj)
```

```python
output.to_csv('Bbolt_output.csv', sep='\t', encoding='utf-8')
output.to_excel("Bbolt_output.xlsx")  
output
```
